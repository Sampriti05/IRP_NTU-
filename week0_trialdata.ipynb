```python
# importing the relevant libraries needed 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# importing sample data
data = pd.read_csv('sample_dataset_IRP.csv')
data.head(n = 10)
```

```python
#plot a line graph 
x = data['Index']
y = data['Value']

plt.plot(x,y)
```

```python
# storing the values as a list to loop through in next step
gradient1 = list()
gradient2 = list()
values = data.iloc[:,1].tolist()
values
```

```python
# loop to calculate gradient (first derivative) 
for i in range(len(values) - 1):  
    grad = (values[i+1] - values[i]) / 2
    gradient1.append(grad)

print(gradient1)
```

```python
# loop to calculate gradient (second derivative) 
for i in range(len(gradient1) - 1):  
    grad = (gradient1[i+1] - gradient1[i]) / 2
    gradient2.append(grad)

print(gradient2)
```

```python
#to find the peak points second derivative = 0 
#for numbers that are consecutive defines single point , there should be 19 changes and 13 obvious peaks 
#including the starting point based on the graph above 
#to equate to 0 is not enough hence a trial and error is needed to change the atol based on the actual data points
#in the below method instead 
peak2 = list()
for index, val in enumerate(gradient2):
    if val == 0.0:
        peak2.append(index)
        
    
print(peak2)
```

```python
#to find the peak points second derivative ~ 0 estimate to ensure no changes are missed out 
peak = list()
for index, val in enumerate(gradient2):
    if np.isclose(val, 0.0, atol=1e-10):
        peak.append(index)
        
    
print(peak)    
```
