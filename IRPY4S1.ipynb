```python
#LOESS/LOWESS method for smoothing data
import statsmodels.api as sm


x = relevant_data_1["exp_time"].values
y = relevant_data_1["volume_acid/base_added"].values

# Fit LOWESS
lowess = sm.nonparametric.lowess
smoothed1 = lowess(y, x, frac=0.1)

# Bootstrap confidence intervals
n_boot = 100
boot_preds = np.zeros((n_boot, len(x)))

for i in range(n_boot):
    sample_idx = np.random.choice(len(x), len(x), replace=True)
    boot_x = x[sample_idx]
    boot_y = y[sample_idx]
    boot_fit = lowess(boot_y, boot_x, frac=0.3, return_sorted=False)
    boot_preds[i,:] = boot_fit

# Conf Interval bounds (still working on this)
lower = np.percentile(boot_preds, 2.5, axis=0)
upper = np.percentile(boot_preds, 97.5, axis=0)

# Plot
plt.scatter(x, y, alpha=0.5)
plt.plot(smoothed1[:,0], smoothed1[:,1], 'r-', label="LOWESS")
plt.fill_between(x, lower, upper, color="red", alpha=0.2, label="95% CI")
plt.legend()
plt.show()
```
